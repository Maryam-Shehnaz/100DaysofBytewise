```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('titanic .csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Check for any missing values 
print("Checking for missing values:")
df.isnull().sum()
```

```python
# impute age with median
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
# Drop the Cabin column
df.drop(columns=['Cabin'], inplace=True)
```

```python
# Impute  Embarked with the mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Check for duplicate rows
duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")
print(duplicate_rows)
```

Exercise 1: Create a simple line plot using Matplotlib to display the trend of a list of values over time.

```python
import matplotlib.pyplot as plt

values = [10, 12, 15, 14, 20, 22, 25, 28, 30]
time = list(range(1, len(values) + 1)) 

plt.figure(figsize=(10, 5))
plt.plot(time, values, marker='o')
plt.title('Trend of Values Over Time')
plt.xlabel('Time')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

Exercise 2: Plot a bar chart using Matplotlib to show the frequency of different categories in a dataset.

```python
# Calculate frequencies of each embarkation point
frequencies = df['Embarked'].value_counts()
categories = frequencies.index.tolist()
frequencies = frequencies.values.tolist()
```

```python
# Print
print("Categories:", categories)
print("Frequencies:", frequencies)
```

```python
# Create the bar chart
plt.figure(figsize=(10, 5)) 
plt.bar(categories, frequencies)  
plt.title('Frequency of Different Embarkation Points')  
plt.xlabel('Embarkation Point') 
plt.ylabel('Frequency')  
plt.xticks(categories)  
plt.show()  
```

Exercise 3: Create a scatter plot using Matplotlib to visualize the relationship between two variables in a dataset.

```python
age = df['Age']
fare = df['Fare']

# Create the scatter plot
plt.figure(figsize=(10, 5))  
plt.scatter(age, fare, alpha=0.5)  
plt.title('Scatter Plot of Age vs Fare')  
plt.xlabel('Age')  
plt.ylabel('Fare')  
plt.grid(True)  
plt.show()  
```

Exercise 5: Create a box plot using Seaborn to show the distribution of values for different categories in a dataset.


```python
import seaborn as sns

# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Distribution of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

Exercise 6: Plot a heatmap using Seaborn to visualize the correlation matrix of a dataset.


```python
# Select numerical columns 
df_numeric = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]

# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Exercise 7: Use Matplotlib to create a subplot grid that displays multiple charts in a single figure.

```python
# 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Histogram of Ages
axs[0, 0].hist(df['Age'].dropna(), bins=20, color='blue', edgecolor='black')
axs[0, 0].set_title('Histogram of Ages')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Frequency')

# Plot 2: Bar Chart of Passenger Classes
class_counts = df['Pclass'].value_counts()
axs[0, 1].bar(class_counts.index, class_counts.values, color='green', edgecolor='black')
axs[0, 1].set_title('Bar Chart of Passenger Classes')
axs[0, 1].set_xlabel('Passenger Class')
axs[0, 1].set_ylabel('Count')

# Plot 3: Scatter Plot of Age vs. Fare
axs[1, 0].scatter(df['Age'], df['Fare'], alpha=0.5)
axs[1, 0].set_title('Scatter Plot of Age vs. Fare')
axs[1, 0].set_xlabel('Age')
axs[1, 0].set_ylabel('Fare')

# Plot 4: Box Plot of Fare by Passenger Class
axs[1, 1].boxplot([df[df['Pclass'] == 1]['Fare'].dropna(),
                   df[df['Pclass'] == 2]['Fare'].dropna(),
                   df[df['Pclass'] == 3]['Fare'].dropna()],
                  labels=['1st Class', '2nd Class', '3rd Class'])
axs[1, 1].set_title('Box Plot of Fare by Passenger Class')
axs[1, 1].set_xlabel('Passenger Class')
axs[1, 1].set_ylabel('Fare')

plt.tight_layout()

plt.show()
```

Exercise 8: Customize the appearance of a Seaborn plot by changing the color palette, adding titles, and modifying axis labels.

```python

# Set a custom color palette
custom_palette = sns.color_palette("Set2")

# Create the box plot with the custom palette
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=df, palette=custom_palette)

plt.title('Distribution of Fare by Passenger Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=14)
plt.ylabel('Fare', fontsize=14)

# Customize tick parameters
plt.xticks(ticks=[0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'], fontsize=12)
plt.yticks(fontsize=12)

plt.grid(True, linestyle='--', alpha=0.6)

# Display
plt.show()
```

Exercise 9: Create a violin plot using Seaborn to visualize the distribution of a dataset across different categories.

```python
plt.figure(figsize=(10, 6))

# Create the violin plot
sns.violinplot(x='Pclass', y='Fare', data=df, palette='muted')

plt.title('Distribution of Fare by Passenger Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=14)
plt.ylabel('Fare', fontsize=14)

# Customize tick parameters
plt.xticks(ticks=[0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'], fontsize=12)
plt.yticks(fontsize=12)

# Display
plt.show()
```

Exercise 10: Combine Matplotlib and Seaborn to create a complex visualization, such as overlaying a KDE plot on a histogram.

```python
plt.figure(figsize=(10, 6))

# Create the histogram
plt.hist(df['Age'].dropna(), bins=30, color='lightgreen', edgecolor='black', alpha=0.7, label='Histogram')

# Overlay the KDE plot
sns.kdeplot(df['Age'].dropna(), color='red', linewidth=2, label='KDE')

plt.title('Distribution of Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.legend()

# Display 
plt.show()
```
